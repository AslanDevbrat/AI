# Calculus 

### The Calculus of Infinitesimals

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) # start, finish, n points
```

If $y = x^2 + 2x + 2$: 

```python
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x,y)
```

* There are no straight lines on the curve. 
* If we zoom in _infinitely_ close, however, we observe curves that _approach_ lines. 
* This enables us to find a slope $m$ (tangent) anywhere on the curve, including to identify where $m = 0$: 

```python
fig, ax = plt.subplots()
ax.set_xlim([-2, -0])
ax.set_ylim([0, 2])
_ = ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-1.5, -0.5])
ax.set_ylim([0.5, 1.5])
_ = ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-1.1, -0.9])
ax.set_ylim([0.9, 1.1])
_ = ax.plot(x,y)
```
